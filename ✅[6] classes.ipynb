## Student Class
Define a student class with the followings:
- A first_name attribute.
- A last_name attribute.
- An id attribute.
- A courses attribute that stores all passed courses and the grades in dictionary.
- An \_\_init\_\_ method.
- A \_\_str\_\_ method.
- A get\_gpa method.
- A set\_course_grade method.

```python
class student:
    def __init__(self, name, last, age, s_id, courses):
        self.name = name
        self.last = last
        self.age = age
        self.s_id = s_id
        self.courses = courses
        
    def get_gpa(self):
        grades = self.courses.values()
        
        return sum(grades)/len(grades)
    
    def set_course_grade(self, new_course):
        self.courses.update(new_course)
    
    def __str__(self):
        return "Student {} is {} years old he's id is: {} and hes grades are: {}".format(self.name, self.age, self.s_id, self.courses)
    
    @classmethod
    def from_string(cls, student_str):
        name, last, age, id, courses = student_str.split('-')
        return cls(name, last, age, id, courses)

p1 = student('Jorge', 'Sisco', '27', '21313434', {'math':10, 'history':7, 'Chemistry': 5})

new_student_1 = 'Jesus-Lopez-27-12134434-{"math":10, "history":7, "Chemistry": 5}'


# new_course = {'comunication': 10}
# p1.set_course_grade(new_course)
# p1.get_gpa()

p2 = student.from_string(new_student_1)

print(p2)
```

## University Class
Define a university class with the followings:
- A name attribute.
- A students attribute that stores a list of students.
- An \_\_init\_\_ method.
- A \_\_str\_\_ method.
- A register\_student method.
- A graduate\_student method.

```python
class university:
    
    def __init__(self, U_name, students):
        self.U_name = U_name
        self.students = students
        
    def __str__(self):
        return "In the {} we have {} students".format(self.U_name, len(self.students))

student_1 = student('Jorge', 'Sisco', '27', '21313434', {'math':10, 'history':7, 'Chemistry': 5})
student_2 = student('Hasan', 'Lopez', '27', '21313434', {'math':10, 'history':7, 'Chemistry': 5})

list_students = [student_1, student_2]

u = university('Gisma', list_students)

print(u)
```
